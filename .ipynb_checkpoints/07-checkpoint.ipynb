#### Overfitting (과대적합), Underfitting (과소적합)

```python
import pandas as pd

df = pd.read_csv('test1_modify.csv')

label = 'count'
features = ['season', 'holiday', 'workingday', 'weather',
            'temp', 'atemp', 'humidity', 'windspeed',
            'year', 'month', 'day', 'hour', 'dayofweek']

train = df[0::2]
test = df[1::2]
train = train.reset_index()
test = test.reset_index()

X_train, Y_train = train[features], train[label]
X_test, Y_test = test[features], test[label]
```

```python
from sklearn.tree import DecisionTreeRegressor as dt

model_dt = dt()
model_dt.fit(X_train, Y_train)
```

```python
model_dt.score(X_train, Y_train)

# 과적합 (Overfitting)
```

```python
model_dt.score(X_test, Y_test)

# 일반화 실패
```

#### 다양한 회귀문제의 평가지표들
- R (root), M (mean), S (squared), L (log), P (pecrentile), A (absolute), E (error)
- MAE (Mean Absolute Error)
- MSE (Mean Squared Error)
- RMSE (Root Mean Squared Error)
- RMSLE (Root Mean Squared Log Error)
- MAPE (Mean Absolute Pecrentile Error)

```python
import numpy as np

err1 = np.log(2) - np.log(1)
err2 = np.log(1002) - np.log(1001)

print(err1)
print(err2)

# 로그오차 : 작은 값들끼리의 차이와 큰 값들끼리의 차이에 차등을 두기 위함
```

```python
from sklearn.metrics import mean_absolute_error as MAE
```

```python
MAE(Y_test, model_rf.predict(X_test))
```

```python
MAE(Y_test, model_dt.predict(X_test))
```

#### feature selection
- 모든 feature를 다 사용하지 않고 골라서 사용함
- 과적합을 방지하기 위해 사용
- 성능을 상승시켜주기 위해 사용
- 방법론 3가지 : filter, wrapper, embed

#### Filter
- 기초통계량을 기반으로 feature 간 관계에서 높은 상관계수를 가지는 feature를 사용하는 방법
- Filter 방식이 유효한 경우는 분야마다 다름

- Wrapper 방법 : forward selection, backward Elimination, stepwise selection

```python
from sklearn.feature_selection import RFE
```

```python
label = 'count'
features = ['season', 'holiday', 'workingday', 'weather',
            'temp', 'atemp', 'humidity', 'windspeed',
            'year', 'month', 'day', 'hour', 'dayofweek']

train = df[0::2]
test = df[1::2]
train = train.reset_index()
test = test.reset_index()

X_train, Y_train = train[features], train[label]
X_test, Y_test = test[features], test[label]

model = rf()
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
```

```python
selector = RFE(model)
selector.fit(X_train, Y_train)
```

```python
rank = selector.ranking_
```

```python
list(zip(features, rank))
```

```python
rfe_df = pd.DataFrame()
rfe_df['feature'] = features
rfe_df['rank'] = rank

rfe_df.sort_values(by='rank')
```

#### Embed
- feature들의 의사결정에 사용된 정도가 있는 경우
- Tree 계열이나 Regularizer가 내부에 있는 경우
- 알고리즘의 공식문서에 Lasso, L1 표현이 있는 경우

```python
impo = model.feature_importances_
impo
```

```python
# 시각화 (내림차순)

embed_df = pd.DataFrame()
embed_df['feature'] = features
embed_df['importance'] = impo

embed_df = embed_df.sort_values(by='importance', ascending=False)
embed_df
```

```python
sns.barplot(data=embed_df, x='importance', y='feature')
```
