#### DataFrame 합치기 (병합하기)

#### concat

```python
import pandas as pd
```

```python
dic = {'가':['가1', '가2', '가3'],
      '나':['나1', '나2', '나3'],
      '다':['다1', '다2', '다3'] }

df1 = pd.DataFrame(dic)
df1
```

```python
dic = {'나':['나1', '나2'],
      '다':['다1', '다2'],
      '라':['라1', '라2'],
      '마':['마1', '마2'] }

df2 = pd.DataFrame(dic)
df2
```

```python
pd.concat( [df1, df2] )
```

```python
pd.concat( [df1, df2], axis=1 )
```

```python
pd.concat( [df1, df2], join='inner' )
```

#### merge
- column은 전부다 생성 (합집합 = outer)
- row는 중복되는 부분만 생성 (교집합 = inner)

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

## 데이터 분석 방법론
- 전통적인 데이터 분석 방법론 : KDD, CRISP-DM

#### CRISP-DM
- Cross-Industry Standard Process - Data Mining

## 1. Business Understanding
- 다양한 분야의 목적과 요구사항을 이해하는 단계
- 업무 목적, 상황 파악, 목표 설정, 프로젝트 계획 수립
- 해당 분야의 종사자들이 원하는 방향 파악
- 분석의 방향성 및 목표 설정!

## 2. Data Understanding
- 필요한 데이터를 수집하고, 데이터의 속성을 이해하는 단계
- 데이터의 기술적인 분석, EDA, 데이터 품질 확인

#### EDA (Exploratory Data Analysis) : 탐색적 데이터 분석
- 시각화를 통해 데이터에서 유효한 패턴(인사이트, 통찰력)을 찾음
- 그래프, 통계적인 방법 -> 자료를 직관적으로 바라보는 과정
- Business Understanding 단계에서 설정한 방향성이 맞는지 여부 확인

1. 데이터 출처와 주제에 대한 이해
2. 데이터 크기 확인
3. 데이터 구성 요소(feature) 확인
4. 탐색한 데이터의 시각화

#### 데이터 품질
- 결측치
- 일관성
- 형태
- 정보

## 3. Data Proparation
- 분석 기법에 알맞은 데이터를 준비하는 단계
- 모델링을 대비하는 단계
- 알고리즘 넣을 데이터를 '적절하게' 준비 

#### Machine Learning
- 학습 방식에 따라 지도 학습, 비지도 학습으로 나뉨
- 데이터셋 분리(split) : 학습용(train) / 시험용(test)

## 4. Modeling
- 다양한 모델링 기법과 알고리즘을 선택하는 단계
- 모델의 성능 평가, overfitting 되었는지 확인

#### Overfitting (과적합) ↔ Underfitting (과소적합)
- 얼마만큼 학습을 시킬 것인지 정하기 위해 확인

#### 1) Model Planning
- 속도, 예측, 해석력을 고려해서 알고리즘 후보군 선정
- 내부 계산이 복잡할수록 속도↓, 예측력↑, 해석력↓
- 기술통계 분야 알고리즘은 속도와 해석력이 우수함

#### 2) Model Building
- 채점할 때 관점에 따라 평가지표가 달라짐
- ex1) 정답 : 1, 예측 : 2 --> 오차 : 1
- ex2) 정답 : 1001, 예측 : 1002 --> 오차 : 1

## 5. Evaluation
- 결과 토의

## 6. Deployment
- 완성된 모델을 실제 업무에 적용하기 위한 단계
