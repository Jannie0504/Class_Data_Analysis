#### Data Preparation
- train / test split → row 분리
- features / label split → column 분리

```python
import pandas as pd

df = pd.read_csv('test1_modify.csv')
df.head()
```

```python
df.columns
```

```python
# feature / label split

label = 'count'
features = ['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'year', 'month', 'day', 'hour', 'dayofweek']
```

```python
# train / test split

train = df[0::2]
test = df[1::2]
```

```python
train
```

```python
test
```

```python
# train / test split

train = train.reset_index()
test = test.reset_index()
```

```python
train
```

```python
test
```

```python
# train, test 각각 feature, label 을 생성
# 모델링을 할 때는 이렇게 4개의 값이 필요함

X_train, Y_train = train[features], train[label]
X_test, Y_test = test[features], test[label]
```

#### Modeling
- 지도학습
1. classification (분류모델) - 객관식, 경우의 수 (Classifier)
2. regression (회귀모델) - 주관식, 숫자 (Regressor)
3. 예외 : Logistic Regression (분류 알고리즘)

```python
from sklearn.ensemble import RandomForestRegressor as rf
```

```python
model = rf()
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
model.score(X_train, Y_train)

# train 값을 넣어도 100%가 나오지 않는다
# Overfitting 때문에 완벽하게 학습하지 않는다.
# 완벽하게 학습하면 test 할 때, 문제 발생
```

#### 결정계수 (R-squared, R^2)
- 회귀 분석이 얼마나 잘 되었는지 평가하는 지표
- 0 ~ 1 사이의 값을 갖고, 1에 가까울수록 예측력이 좋다고 판단

- SSE (오차 제곱 합) = (actual - pred)^2 의 합
- SSR (회귀 제곱 합) = (pred - avg)^2 의 합
- SST (전체 제곱 합) = (actual - avg)^2 의 합
- R^2 = 1 - (SSE / SST) = SSR / SST
- target에 평균 정도를 예측하는 수준이라면 R^2는 0에 가까워지고, 예측이 target에 가까워지면 1에 가까운 값이 됨

```python
y_pred = model.predict(X_test)
y_pred
```
