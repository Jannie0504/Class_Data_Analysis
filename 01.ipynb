## Data Frame
- 데이터가 틀에 맞춰서 저장된 형태
- 테이블, 스프레드 시트, 엑셀 등 분야마다 다양하게 불림
- 가로(row, 행), 세로(column, 열)
- 한 row 안의 데이터는 Dictionary처럼 저장
- 한 column 안의 데이터는 List처럼 저장
- 한 DataFrame은 row들이 List처럼 저장

```python
import pandas as pd
```

```python
all_list = []

all_list.append({'data':'2023-01-01', 'day':'Sun'})
all_list.append({'data':'2023-01-02', 'day':'Mon'})
all_list.append({'data':'2023-01-03', 'day':'Tue'})

all_list
```

```python
pd.DataFrame(all_list)
```

```python
name = ['Kim', 'Lee', 'Park', 'Choi', 'Lim']
age = [31, 59, 17, 44, 22]

DataSet = list(zip(name, age))
DataSet
```

```python
pd.DataFrame(DataSet, columns=['name', 'age'])
```
