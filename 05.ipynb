## CRISP-DM 기반의 데이터 분석
- 출처 : https://ride.capitalbikeshare.com/system-data
- Maryland 주 Washington D.C에서 자전거 공유 회사가 모은 데이터셋

```python
import pandas as pd

df = pd.read_csv('test1.csv')
print(df.shape)
df.head()
```

#### Business Understanding
- 공유 자전거 서비스 업계의 도메인을 이해
- 자전거라는 상품을 대여
- 대여되지 않은 자전거의 개수가 적을수록 사업이 잘 된다고 판단
- 분석 방향성 : 수요량 예측, 서비스 이용량 증감 파악

```python
df.columns
```

#### Data Understanding
- 2011 ~ 2012년도까지 2년치, 시간대별로 저장
- 날짜 정보 : 'datetime', 'season', 'holiday', 'workingday'
- 날씨 정보 : 'weather', 'temp', 'atemp', 'humidity', 'windspeed'
- 고객의 이용 정보 : 'casual', 'registered', 'count'

```python
sample = '2011-01-01 00:00:00'
sample = sample.split()
sample
```

```python
year = sample[0].split('-')[0]
month = sample[0].split('-')[1]
day = sample[0].split('-')[2]

hour = sample[1].split(':')[0]
minute = sample[1].split(':')[1]
sec = sample[1].split(':')[2]

print(year, month, day, hour, minute, sec)
```

```python
df['year'] = df['datetime'].apply(lambda x : int(x.split()[0].split('-')[0]))
df['month'] = df['datetime'].apply(lambda x : int(x.split()[0].split('-')[1]))
df['day'] = df['datetime'].apply(lambda x : int(x.split()[0].split('-')[2]))
df['hour'] = df['datetime'].apply(lambda x : int(x.split()[1].split(':')[0]))
df.head()
```

```python
sample = '2011-01-01 00:00:00'

parsed = pd.to_datetime(sample)
parsed.dayofweek

# 0:Mon ~ 6:Sun
```

```python
df['dayofweek'] = df['datetime'].apply(lambda x : pd.to_datetime(x).dayofweek)
df.head()
```

#### EDA (탐색적 데이터 분석)

```python
import seaborn as sns

sns.scatterplot(data=df, x='registered', y='casual')
```

```python
sns.scatterplot(data=df, x='registered', y='casual', hue='season')
```

```python
# Q) registered와 casual은 서비스 이용 목적이 다를 것으로 예상된다.
#    이 예상을 뒷받침할 수 있는 그래프를 그려보세요.

sns.scatterplot(data=df, x='registered', y='casual', hue='workingday')

# workingday = 1 --> 일하는 날에는 등록 회원(registered)이 많이 이용함
# workingday = 0 --> 일하지 않는 날에는 비등록 회원(casual)이 많이 이용함
# 등록/비등록 회원의 이용 빈도의 차이는 일하는 날의 유무이다.
```

#### 데이터를 표현 → 문장 도출 → 의미가 있어야 함

```python
sns.lineplot(data=df, x='month', y='registered')
```

```python
# Q) 월별 등록/비등록 회원의 이용량의 편차는 어떠한가?

sns.lineplot(data=df, x='month', y='casual')
```

```python
# Q) 시간에 따른 패턴을 확인해보자.

sns.lineplot(data=df, x='hour', y='registered')
sns.lineplot(data=df, x='hour', y='casual')

# hour = 7 정도 & hour = 17 정도 --> 이용자가 많음
# 출퇴근 시간에 이용자가 많다.
```

```python
# Q) 시간별 workingday 유무에 따른 차이가 있는가?

sns.lineplot(data=df, x='hour', y='registered', hue='workingday')

# workingday = 1 --> 일하는 날에는 출퇴근 시간에 많이 이용함
# workingday = 0 --> 일하지 않는 날에는 아침시간보다는 점심시간부터 초저녁에 많이 이용함
# workingday 유무에 따라 이용하는 시간대 차이가 있다.
```

```python
sns.lineplot(data=df, x='hour', y='casual', hue='workingday')

# workingday = 1 --> 일하는 날에 이용량이 더 많음
# workingday = 0 --> 일하지 않는 날에는 이용량이 현저히 적음
# workingday 유무에 따라 이용하는 양이 다르다.
```

```python
sns.lineplot(data=df, x='dayofweek', y='registered')
sns.lineplot(data=df, x='dayofweek', y='casual')

# 회원/비회원에 따라 주로 이용하는 요일이 다름
# 회원들은 주로 평일에 많이 이용하고, 비회원들은 주로 주말에 더 많이 이용함
```
